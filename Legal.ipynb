```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt

import seaborn as sns
%matplotlib inline
```

```python
df=pd.read_csv('Case_files.csv')
df.head()
```

```python
df.drop('Unnamed: 0',axis=1,inplace=True)
```

```python
df['Brief Fact Summary'] = df['Brief Fact Summary'].astype(str).str.lower()
df['Synopsis of Rule of Law'] = df['Synopsis of Rule of Law'].astype(str).str.lower()
df['Facts'] =df['Facts'].astype(str).str.lower()
df['Issue'] = df['Issue'].astype(str).str.lower()
df['Held'] = df['Held'].astype(str).str.lower()
df['Discussion'] = df['Discussion'].astype(str).str.lower()#.astype(str).str.split()
```

```python
import string
def remove_punctuations(text):
    for punctuation in string.punctuation:
        text = text.replace(punctuation, '')
    return text



df["a"] = df['Brief Fact Summary'].apply(remove_punctuations)
df["b"] = df['Synopsis of Rule of Law'].apply(remove_punctuations)
df["c"] = df['Facts'].apply(remove_punctuations)
df["d"] = df['Issue'].apply(remove_punctuations)
df["e"] = df['Held'].apply(remove_punctuations)
df["f"] = df['Discussion'].apply(remove_punctuations)
df.drop(columns=['Brief Fact Summary','Synopsis of Rule of Law','Facts','Issue','Held','Discussion'],inplace=True)
```

```python
df.head()
```

```python
df['Held'] = ''
positive=0
negative=0
for i in range(len(df)):
    for j in range(len(df['e'][i])):
        if (df['e'][i][j:j+3]=='yes'):
            positive+=1
            df['Held'][i]=1
            break
        elif (df['e'][i][j:j+2]=='no'):
            negative+=1
            df['Held'][i]=0
            break
        else:
            df['Held'][i]=df['e'][i]
```

```python
df['Brief Fact']=df['a'].apply(lambda x: [stemmer.stem(item) for item in x if item not in stop])
df['Synopsis of Rule of Law']=df['b'].apply(lambda x: [stemmer.stem(item) for item in x if item not in stop])
df['Facts']=df['c'].apply(lambda x: [stemmer.stem(item) for item in x if item not in stop])
df['Issue']=df['d'].apply(lambda x: [stemmer.stem(item) for item in x if item not in stop])
df['Discussion']=df['f'].apply(lambda x: [stemmer.stem(item) for item in x if item not in stop])
df.drop(columns=['a','b','c','d','f'],inplace=True)
```

```python
import io
#df = pd.read_csv(io.BytesIO(uploaded['Co_Casefile.csv']))
df = pd.read_csv('Co_Casefile.csv')
df.head()
```

```python
df['Facts']=df['Facts'].str.replace('\d+', '')
df['Issue']=df['Issue'].str.replace('\d+', '')
df.tail(10)
```

```python
df.drop('Unnamed: 0',axis=1,inplace=True)
```

```python
from keras.preprocessing.sequence import pad_sequences
from keras.preprocessing.text import Tokenizer
tokenizer = Tokenizer(num_words=20000, lower=True,split=' ')

tokenizer.fit_on_texts(df['Facts'].values)
tokenizer.fit_on_texts(df['Issue'].values)
#print(tokenizer.word_index)  # To see the dicstionary
X1 = tokenizer.texts_to_sequences(df['Facts'])
X2  = tokenizer.texts_to_sequences(df['Issue'])
X1 = pad_sequences(X1)
X2 = pad_sequences(X2)
```

```python
X = np.concatenate(([X1,X2]),axis = 1)
y=df['Held']
```

```python
y = np.array(y).reshape(y.shape[0],1)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.10, random_state = 37,shuffle=True)
print(X_train.shape,y_train.shape)
print(X_test.shape,y_test.shape)
```

```python
from sklearn import model_selection, naive_bayes, svm
from sklearn.metrics import accuracy_score,confusion_matrix
```

```python
nv=naive_bayes.MultinomialNB()
nv.fit(X_train,y_train)
# predict the labels on validation dataset
predictions_NB = nv.predict(X_test)
# Use accuracy_score function to get the accuracy
print("Naive Bayes Accuracy Score -> ",accuracy_score(predictions_NB, y_test)*100)
```

```python
print(accuracy_score(y_train, nv.predict(X_train))*100)
```

```python
pn=confusion_matrix(predictions_NB,y_test)
xn=pn
xn
```

```python
print("Naive Bayes:\n")
import seaborn as sns
plt.figure(figsize=(10,6))
print(sns.heatmap(pn, fmt='',annot=True,cmap='viridis'))
```

```python
group_names = ['True Neg','False Pos','False Neg','True Pos']
group_percentages = ["{0:.2%}".format(value) for value in
                     xn.flatten()/np.sum(xn)]
labels = [f"{v1}\n\n{v3}" for v1, v3 in
          zip(group_names,group_percentages)]
labels = np.asarray(labels).reshape(2,2)
plt.figure(figsize=(13,5))
sns.heatmap(xn, annot=labels, fmt='', cmap='Blues')
```

```python
# Classifier - Algorithm - SVM
# fit the training dataset on the classifier
SVM = svm.SVC()
SVM.fit(X_train,y_train)
# predict the labels on validation dataset
predictions_SVM = SVM.predict(X_test)
# Use accuracy_score function to get the accuracy
print("SVM Accuracy Score -> ",accuracy_score(predictions_SVM, y_test)*100)
```

```python
print(accuracy_score(y_train, SVM.predict(X_train))*100)
```

```python
xs=confusion_matrix(predictions_SVM,y_test)
ps=xs
ps
```

```python
print("SVM:\n")
import seaborn as sns
plt.figure(figsize=(10,6))
print(sns.heatmap(ps, fmt='',annot=True,cmap='viridis'))
```
